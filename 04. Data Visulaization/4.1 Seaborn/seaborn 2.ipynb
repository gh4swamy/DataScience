```python
import numpy as np, pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'

tips = sns.load_dataset('tips')
tips.head(2)
```

# Historams 

```python
#Histogram of univarite items i.e total _bill and Tips -- numeric variables 

sns.distplot(tips['total_bill'], kde = True, color = 'indianred', bins = 20, )
```

```python
sns.distplot(tips['tip'], kde = False, color = 'indianred', bins = 10)
```

# Joint Plots

```python
# Joint plot of two  distrubution univarite variables suhc as tOtal Bill and Tips --along with scatter plot 
sns.jointplot(x='total_bill', y = 'tip', data = tips)
```

# Pairplot

```python
sns.pairplot(tips, hue = 'smoker', palette = 'coolwarm')
```

# rug plot

```python
sns.rugplot(tips['total_bill'])
```

# KDE plot

```python
sns.kdeplot(tips['total_bill'])
```

# Categorical Plots

```python
tips.head(3)
```

```python
sns.barplot(x= 'sex', y ='total_bill' , data = tips, estimator = np.std)
```

```python
sns.countplot(x='day', data = tips)
```

# Box Plots

```python
sns.boxplot(x='day', y ='total_bill', data = tips)
```

# Violine Plot

```python
sns.violinplot(x='day', y ='total_bill', data = tips, hue = 'smoker', split= True, cut = 3 )
```

# stripplot

```python

sns.stripplot(x='day', y = 'tip', data = tips, jitter= True, hue = 'sex', split = True)
```

# Swarm Plot 

```python
sns.violinplot(x = 'sex', y = 'total_bill', data = tips, hue = 'smoker', split = True)
sns.swarmplot(x = 'sex', y = 'total_bill', data = tips, hue = 'smoker', split = True, palette='husl')
```

# Factor plot

```python
sns.factorplot(x = 'day', y = 'total_bill', data = tips, kind = 'swarm')
```

# Matrix Plots 

## Heatmaps 

```python
flts= sns.load_dataset('flights')
flts.head()
```

```python
df = flts.pivot_table(index= 'month', columns='year', values='passengers')
df
```

```python
tc = tips.corr()
tc
```

```python
sns.heatmap(tc, annot = True, lw=3, cmap = 'coolwarm')
```

```python
plt.figure(figsize = (15, 10))
sns.set_context('poster')
sns.heatmap(df, cmap = 'coolwarm', linecolor = 'orange', linewidth = 3)
```

# cluser map

```python
sns.clustermap(df, cmap = 'coolwarm')
```

# Regression Plots

```python
plt.figure(figsize= (15, 15))
sns.lmplot(x= 'total_bill', y = "tip", data =tips, hue = 'sex', markers = ['o', 'v'], scatter_kws = {'s': 100})
```

```python
sns.lmplot(x= 'total_bill', y = "tip", data =tips, col= 'day', row = 'sex', hue = 'smoker', aspect = 0.8, size = 5)
```
