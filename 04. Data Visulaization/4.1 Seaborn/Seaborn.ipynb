```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline 
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.barplot(x= 'sex', y= 'total_bill', data =tips, estimator= np.std)
```

```python
sns.countplot(x= 'sex' , data = tips )
```

```python
sns.boxplot(x = 'day',y = 'total_bill', data = tips, hue = 'smoker')
```

```python
sns.violinplot(x = 'day',y = 'total_bill', data = tips, hue = 'sex', split = True)
```

```python
sns.stripplot(x = 'day', y = 'total_bill', data = tips, jitter = True, hue = 'sex', split= True )
```

```python
sns.swarmplot(x= 'day', y='total_bill' , data = tips)
```

```python
sns.violinplot(x= 'sex', y='total_bill' , data = tips)
sns.swarmplot(x= 'sex', y='total_bill' , data = tips, color = 'green', hue = 'smoker')
```

```python
sns.factorplot(x= 'sex', y='total_bill' , data = tips, kind = 'bar')
```

# Matrix  Plots 

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
tc = tips.corr()
```

```python
sns.heatmap(tc, annot = True, cmap = 'coolwarm')
```

```python
fp = flights.pivot_table(index= 'month', columns= 'year', values = 'passengers')
```

```python
sns.heatmap(fp, cmap = 'magma', linecolor='white', linewidth = 3, )
```

```python
sns.clustermap(fp)
```

# Regression Plots 

```python
tips.head()
```

```python
sns.lmplot(x= 'total_bill', y = 'tip', data = tips, hue = 'sex', markers = ['*', 'v'],
          scatter_kws = {'s': 100})
```

```python
sns.lmplot(x='total_bill', y = 'tip', data = tips, col = 'day', hue = 'sex', 
          aspect= 0.6, size = 8)
```

# Grids

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(iris, hue = 'species')
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)

```

```python
tips.head()
```

```python
g = sns.FacetGrid(data = tips, col = 'time', row ='smoker', hue = 'sex')
g.map(plt.scatter, 'total_bill', 'tip')
```
