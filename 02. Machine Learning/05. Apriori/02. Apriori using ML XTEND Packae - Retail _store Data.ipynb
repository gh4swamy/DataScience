## France Data with mlxtend package


```python
from mlxtend.frequent_patterns import apriori, association_rules
fr = pd.read_csv('france.csv')
fr.head(20)
```

```python
fr.shape
```

```python
fr_group = fr.groupby('InvoiceNo').agg({'Quantity':np.mean})
len(fr_group) # hence no. o funioue trancations are 603
```

```python
basket_matrix = fr.pivot_table(index='InvoiceNo', columns='Description', values='Quantity', fill_value=0)
basket_matrix.head()
```

```python
basket_matrix.columns
```

```python
def encode_unit(x):
    if x<=0:
        return 0
    elif x>=1:
        return 1
basket_matrix= basket_matrix.applymap(encode_unit)
basket_matrix.drop('POSTAGE', inplace=True, axis=1)
        
```

```python
frequent_items = apriori(basket_matrix, min_support=0.05, use_colnames=True)
frequent_items.head()
frequent_items.sort_values('support', ascending=False).head(10)
```

```python
frequent_items.shape
```

```python
rules = association_rules(frequent_items, min_threshold=1, metric='lift')
rules.tail()
```
