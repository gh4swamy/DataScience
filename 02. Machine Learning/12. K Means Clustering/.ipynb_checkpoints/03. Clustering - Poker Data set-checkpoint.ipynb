```python
import pandas as pd
import numpy as np

import matplotlib.pylab as plt
from sklearn.model_selection import train_test_split
from sklearn import preprocessing

%matplotlib inline

train = pd.read_csv('KMean_poker_train.csv', header=None, names=['S1', 'C1', 'S2', 'C2', 'S3', 'C3','S4', 'C4', 'S5', 'C5', 'CLASS'])
test = pd.read_csv('KMeans_poker_test.csv', header=None, names=['S1', 'C1', 'S2', 'C2', 'S3', 'C3','S4', 'C4', 'S5', 'C5', 'CLASS'])
train.head()
```

```python
test.head()
```

```python
df = train[['S1', 'C1', 'S2', 'C2', 'S3', 'C3','S4', 'C4', 'S5', 'C5']]
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
df = sc.fit_transform(df).astype('float64')
```

```python
df
```

```python
from sklearn.cluster import KMeans
error = []
for i in range(1,11):
    km = KMeans(n_clusters=i, max_iter=300)
    km.fit(df)
    error.append(km.inertia_)
```

```python
plt.plot(range(1,11), error)
plt.show()
```

```python
model = KMeans(n_clusters=2)
model.fit(df)
pred = model.predict(df)
```

```python
from sklearn.decomposition import PCA 
pca_2 = PCA(2) 
plot_columns = pca_2.fit_transform(df)
plt.scatter(x=plot_columns[:,0], y=plot_columns[:,1], c=model.labels_)
plt.xlabel('Canonical variable 1')
plt.ylabel('Canonical variable 2')
plt.title('Scatterplot of Canonical Variables for 2 Clusters')
plt.show() 
```
