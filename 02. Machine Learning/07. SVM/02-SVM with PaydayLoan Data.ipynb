```python
import pandas as pd
df = pd.read_csv('paydayloan_collections.csv')
df.head(3)
```

```python
df['payment'].value_counts()
```

```python
df['payment']=pd.Categorical(df['payment']).codes
```

```python
df.head(3)
```

```python
df.isnull().any()
```

```python
cols = df.columns
print(df.shape)
print(df.columns)

for column in cols:
    if df[column].dtypes=='object':
        #print(column)
        dummy = pd.get_dummies(df[column], prefix=column, drop_first=True)
        df = pd.concat([df, dummy], axis=1)
        df.drop([column], axis=1, inplace=True)
        
        
print("After Dummification")
print(df.shape)
print(df.columns)
```

```python
df.head()
```

```python
x = df.drop('payment', axis=1).values
y = df['payment'].values
```

```python
from sklearn.model_selection import train_test_split
xtr, xts, ytr, yts = train_test_split(x, y , test_size= .3, random_state=0)
```

```python
from sklearn.svm import SVC
svc = SVC(class_weight= 'balanced')

# grid Search 
from sklearn.model_selection import GridSearchCV
params = {'kernel':['rbf', 'linear', 'simoid', 'poly'], 'C':[0.01, 0.1]}

grid = GridSearchCV(svc, param_grid=params, cv=10, scoring='roc_auc')
grid.fit(xtr, ytr)
```
